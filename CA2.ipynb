## Market Basket Analysis

```python
transactions_df = pd.read_csv("transactionsDf.csv")
```

```python
merged = pd.merge(transactions_df, df[["customerName","Category"]], on='customerName', how='left')
merged
```

```python
basket = merged.groupby(["orderNumber", "productName"])["productName"].apply(lambda x: 1).unstack().fillna(0).reset_index().set_index("orderNumber")
basket
```

```python
def encoding(x, true=True, false=False):
    if x <= 0:
        return false
    elif x >= 1:
        return true
    
basket_enc = basket.applymap(encoding)
basket_enc
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(basket_enc, min_support=0.075, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules = rules.sort_values("confidence", ascending=False)
rules.head(20)
```

```python
rules.shape
```

```python
rules.columns = ['Antecedents', 'Consequents', 'Antecedent Support', 'Consequent Support', 'Support', 'Confidence', 'Lift', 'Leverage', 'Conviction', 'Zhangs Metric']
rules['Antecedents'] = rules['Antecedents'].astype(str)
rules['Consequents'] = rules['Consequents'].astype(str)

def remove_unwanted_text(text):
    text = text.replace("frozenset({'", "").replace("'})", "")
    return text
rules['Antecedents'] = rules['Antecedents'].apply(remove_unwanted_text)
rules['Consequents'] = rules['Consequents'].apply(remove_unwanted_text)

rules.to_csv("ruleSet.csv")
```

### Multiple Linear Regression

```python
df = pd.read_csv("salesRegression.csv")
```

```python
df["profit"] = df["priceEach"] - df["buyPrice"]
```

```python
df['orderDate'] = df['orderDate'].apply(lambda x: x[:4])
df["employeeTerritory"] = df["employeeTerritory"].fillna("NA")
df["make"] = ""
mapping = {
    'Ships': 'Ship',
    'Planes': 'Plane',
    'Trucks and Buses': 'Trucks and Buses',
    'Motorcycles': "Motorcycles",  
    'Classic Cars': '',  
    'Vintage Cars': ''  
}
for index, row in df.iterrows():
    product_line = row['productLine']
    product_name = row['productName']
    
    if product_line in mapping:
        if mapping[product_line]:
            # Encode ship, plane, or truck and buses
            df.at[index, 'make'] = mapping[product_line]
        else:
            # For Classic Cars or Vintage Cars, take the second word as the make
            words = product_name.split()
            if len(words) >= 2:
                df.at[index, 'make'] = words[1]
```

```python
df.head()
```

```python
target = "profit"
df[target] = np.log(df[target])

# Print the transformed values
print(df[target])
```

```python
y = df[target]
X = df.drop([target], axis=1)
```

```python
X_cat = ["productName","productLine", "productScale", "customerName", "customerCountry", "customerCity", "employeeName",
           "employeeCity", "employeeCountry", "employeeTerritory", "orderDate", "make"]
X_cont = ["quantityOrdered", "customerCreditLimit", "buyPrice", "MSRP"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

def scaling(X):
    scaler.fit(X)
    X_scaled = scaler.transform(X)
    X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
    return X_scaled


X_scale = X[X_cont]
X_scaled = scaling(X_scale)
X_scaled
```

```python
X_enc_cat = X[X_cat]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)

X_encoded = encoder.fit_transform(X_enc_cat)
column_names = encoder.get_feature_names_out(input_features=X_enc_cat.columns)
X_encoded_df = pd.DataFrame(X_encoded, columns=column_names)
```

```python
X_encoded_df
```

```python
X_scaled
```

```python
mergedf = pd.concat([X_encoded_df, X_scaled], axis=1)
mergedf
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(mergedf, y, test_size=0.2, random_state=111)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
r2
```

```python
from sklearn.metrics import mean_squared_error

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
mse
```

## Lasso Regression

```python
#Lasso
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.001)  
lasso.fit(X_train, y_train)
lasso_coefs = lasso.coef_ 
top_correlated_indices = np.argsort(np.abs(lasso_coefs))[::-1]  
top_correlated_features = X_train.columns[top_correlated_indices]
```

```python
for feature, coef in zip(top_correlated_features, lasso_coefs[top_correlated_indices]):
    print(f"Feature: {feature}\tCoefficient: {coef}")
```

```python
y_pred = lasso.predict(X_test) 

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
